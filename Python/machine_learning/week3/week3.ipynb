```python
# 효율적인 코드 만들기

import numpy as np
import matplotlib.pyplot as plt
import time

t = np.linspace(0, 5, 50000)

y1 = np.exp(-t)
y2 = np.cos(np.pi*t)

start1 = time.time() + 0.000001
y3 = y1*y2
delta_t1 = time.time()-start1
print("1. Operation time: {0} sec".format(delta_t1))

start2 = time.time()
y4 = np.empty((len(t),))
for n in range(len(t)):
    y4[n] = y1[n] * y2[n]
delta_t2 = time.time()-start2
print("2. Operation time: {0} sec".format(delta_t2))
print(delta_t2/delta_t1)

plt.subplot(2,1,1)
plt.plot( t, y3 )
plt.legend(['Case 1'])
plt.subplot(2,1,2)
plt.plot( t, y4 )
plt.legend(['Case 2'])
```

```python
# 사용자 정의 함수

import numpy as np
import matplotlib.pyplot as plt

def op_times_two(x):
    y = 2 * x
    return y

t = np.linspace(0, 2.5, 100)
f = 1
y = np.cos(2*np.pi*f*t)
y2 = op_times_two(y)

plt.plot( t, y, t, y2 )
plt.legend(['Original','Modified'])
plt.grid(True)
```

```python
# 사용자 정의 함수

import numpy as np
import matplotlib.pyplot as plt

def op_times2_divide2(x):
    y1 = 2 * x
    y2 = x / 2
    return y1, y2

t = np.linspace(0, 2.5, 100)
f = 1
y = np.cos(2*np.pi*f*t)

y1, y2 = op_times2_divide2(y)

plt.plot( t, y, t, y1, t, y2 )
plt.legend(['Original','Modified-1','Modified-2'])
plt.grid(True)
```

```python
# 사용자 정의 함수

import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
def g(t):
    return np.sin(np.pi*t)
t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.plot(t1, f(t1), t2, g(t2), 'r-')
plt.show()
```
